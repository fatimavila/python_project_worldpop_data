**For this project, I will define the correlation coefficients as:**

**Strong:** *from 0.75 to 1*

**Moderately strong:** *from 0.50 to 0.75*

*Analysis of the variable life expectancy*

For the variable of "life expectancy" (for both men and women) I observe that it has a strong negative correlation with "Death rate" and a strong positive relation with: "rights" and "health". And a moderately strong positive correlation with: "stability" and "safety".

*Analysis of the variable IQ*

For the variable "IQ" has a strong positive correlation with "life expectancy" (for both male and female) and "health". It has a moderately strong positive correlation with "stability", "rights", "safety", "male/female height", and "education expenditure per inhabitant".  I observe that it has a strong negative correlation with "birth rate" and a moderately strong negative correlation with "daily max temperature".

According to the aforementioned variables that relate to my target variables, these are my hypotheses:


The more healthy a country is, the highest the life expectancy.

The more education expenditure per inhabitant the higher the IQ.


````python
#Exploring the hypotheses:
#Checking coefficient between Male Life Expectancy and Health
import numpy as np
df['male_life_expectancy'].fillna(df['male_life_expectancy'].mean(), inplace=True)
df['health'].fillna(df['health'].mean(), inplace=True)

correlation_matrix = np.corrcoef(df['male_life_expectancy'], df['health'], rowvar=False)
print(correlation_matrix)

#Analyzing Female Life Expectancy and Health
df['female_life_expectancy'].fillna(df['female_life_expectancy'].mean(), inplace=True)
df['health'].fillna(df['health'].mean(), inplace=True)

correlation_matrix = np.corrcoef(df['female_life_expectancy'], df['health'], rowvar=False)
print(correlation_matrix)

#Analyzing education expenditure per inhabitant and IQ
correlation_matrix = df[['education_expenditure_per_inhabitant', 'iq']].corr()

correlation_coefficient = correlation_matrix.loc['education_expenditure_per_inhabitant', 'iq']
print("Correlation coefficient between 'education_expenditure_per_inhabitant' and 'iq':", correlation_coefficient)
````

**Analyzing the first hypothesis: The more healthy a country is, the highest the life expectancy. **

For male life expectancy we observe a positive correlation coefficient of 0.6775. This suggest a moderate positive correlation between both variables: higher male life expectancy is associated with higher values of health.

For female life expectancy we observe a strong positive correlation coefficient of 0.7172. This suggest a strong positive correlation between both variables: higher female life expectancy is associated with higher values of health.

*Analyzing the second hypothesis: the more education expenditure per inhabitant the higher the IQ*

We observe a positive moderate correlation coefficient between both variables  of 0.554. This suggests a moderately positive relationship between education expenditure per inhabitant and IQ. As education spending increases, there tends to be a corresponding rise in average IQ levels among the population.


# How the dataframe was created


This section is not relevant for doing the project, you can ignore it.

In case the dataset needs to be recreated, or if you are a very curious student, this is how it was done (on a local machine, _not_ in Colab):

```python
import pandas as pd
import glob
from functools import reduce

# 1. Download and extract data from
#    https://www.kaggle.com/datasets/madhurpant/world-population-data

# 2. Merge all dataframes
joint_df = reduce(
    lambda df1, df2: df1.merge(df2, on="country", how="outer"),
    [pd.read_csv(path) for path in glob.glob("world-population-data/*")],
)

# 3. Make "country" the index
joint_df.set_index(["country"], inplace=True)

# 4. Save the result
joint_df.to_csv("world-population-data.csv")
```
